```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
import math
import random
```

Consider an Erdos-Renyi network with $N = 3000$ nodes, connected to each other with probability $p = 10^{-3}$.

```python
N = 3000
p = 1e-3
```

The expected number of links is:
$\langle L \rangle = \frac{p N (N - 1)}{2} = \frac{10^{-3} \cdot 3000 \cdot (3000 - 1)}{2}$

```python
L_mean = p * N * (N - 1) / 2
print('The expected number of links is:', L_mean)
```

### In which regime is the network?

```python
k_mean = p * (N - 1)
print('Average degree:', k_mean)
print(p > 1 / N)
```

As $p > 1 / N$ the network is at least in the supercritical regime, which can also be seen by $\langle k \rangle > 1$.

```python
print('Is the network in the connected regime?', p > math.log(N) / N)
```

### Calculate the probability $p_c$ so that the network is at the critical point

We can find this variable by $p_c = \frac{1}{N - 1}$

```python
p_c = 1 / (N - 1)
print('p_c:', p_c)
```

### Given the linking probability $p = 10^{â€“3}$, calculate the number of nodes $N^{cr}$ so that the network has only one component.

We can find this by solving $p = \frac{\ln{N}}{N} \rightarrow N_{rc} > 9118.01 \approx 9119$

```python
N_cr = 9119
```

### For the network in ($d$), calculate the average degree $\langle k_{cr} \rangle$ and the average distance between two randomly chosen nodes $\langle d \rangle$.

```python
k_cr_mean = p * (N_cr - 1)
d_cr_mean = math.log(N_cr) / math.log(k_cr_mean)

print('Average degree:', k_cr_mean)
print('Average distance:', d_cr_mean)
```

### Calculate the degree distribution $p_k$ of this network (approximate with a Poisson degree distribution).

Using $p_k = e^{- \langle k \rangle} \frac{\langle k \rangle^k}{k!}$

```python
def p_k(k): 
    return math.exp(-k_cr_mean) * math.pow(k_cr_mean, k) / math.factorial(k)
```

```python
import numpy as np

k = np.arange(0.0, 20, 1)
values = [p_k(x) for x in k]
plt.plot(k, values)
plt.show()
```

## Paths

```python
N = 200
k_mean = 1.5
p = k_mean / (N - 1)
```

```python
G = nx.erdos_renyi_graph(N, p)
nx.draw(G, node_size=50, edge_color='#009F9F')
```

```python
subgraphs = list(nx.connected_component_subgraphs(G))
GCC = subgraphs[0]

node = random.choice(list(GCC.nodes()))
nodes_two_or_less_away = nx.single_source_shortest_path_length(G, node, cutoff=2)
nodes_two_away = [x for x in nodes_two_or_less_away if nodes_two_or_less_away[x] == 2]

node_colors = []

for n in GCC.nodes():
    if n == node:
        node_colors.append('#000000')
    elif n in nodes_two_away:
        node_colors.append('#ff0000')
    else: 
        node_colors.append('#A0CBE2')

nx.draw(GCC, node_size=100, node_color=node_colors)
```
