4. Read in the texts of the *State of the Union* addresses, using the `state_union` corpus reader. Count occurrences of `men`, `women`, and `people` in each document. What has happened to the usage of these words over time?

````python
from nltk.corpus import state_union

history = {
    'men': [],
    'women': [],
    'people': [],
}
years = []
for filename in state_union.fileids():
    text = nltk.corpus.state_union.words(filename)
    years.append(filename[:4])
    for word in history.keys():
        history[word].append(text.count(word))

plt.subplots(1, 1, figsize=(16, 8))
for key, value in history.items():
    plt.plot(years, value, label=key)

plt.legend(loc='best')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Word count')

plt.show()
````

8. Define a conditional frequency distribution over the Names corpus that allows you to see which *initial* letters are more frequent for males vs. females (cf. 4.4).

````python
from nltk.corpus import names
          
plt.figure(figsize=(16, 4))
nltk.ConditionalFreqDist((fileid, name[0])
                          for fileid in names.fileids()
                          for name in names.words(fileid)).plot()

# Normalizing the distribution
_, _ = plt.subplots(1, 1, figsize=(16, 4))
counter = {}
for fileid in names.fileids():
    counter[fileid] = {}
    for name in names.words(fileid):
        if name[0] in counter[fileid]:
            counter[fileid][name[0]] += 1
        else: 
            counter[fileid][name[0]] = 1
            
    for key, value in counter[fileid].items():
        counter[fileid][key] = counter[fileid][key] / len(names.words(fileid))
    
    plt.plot(counter[fileid].keys(), counter[fileid].values(), label=fileid)
   
plt.legend(loc='best')
plt.xlabel('First letter of name')
plt.ylabel('Frequency of occurrence')
plt.show()
````

17. Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

````python
from nltk.corpus import stopwords

def frequently_occuring_words(text):
    sws = set(stopwords.words('english'))
    filtered_text = [word for word in text if word not in sws]
    
    return sorted(list(FreqDist(filtered_text).items()), 
                  key=lambda x: x[1], reverse=True)[:50]

print(frequently_occuring_words(text1))
````

18. Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

````python
def frequent_bigrams(text):
    sws = set(stopwords.words('english'))
    bigrams = [bigram for bigram in nltk.bigrams(text)
               if bigram[0] not in sws and bigram[1] not in sws]
    return sorted(list(FreqDist(bigrams).items()),
                  key=lambda x: x[1], reverse=True)[:50]
    
print(frequent_bigrams(text2))
````

23.  Zipf's Law: Let $f(w)$ be the frequency of a word $w$ in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipf's law states that the frequency of a word type is inversely proportional to its rank (i.e. $f \times r = k$, for some constant $k$). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.

    - Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipf's law? (Hint: it helps to use a logarithmic scale). What is going on at the extreme ends of the plotted line?
    
    - Generate random text, e.g., using random.choice("abcdefg "), taking care to include the space character. You will need to import random first. Use the string concatenation operator to accumulate characters into a (very) long string. Then tokenize this string, and generate the Zipf plot as before, and compare the two plots. What do you make of Zipf's Law in the light of this?


````python
def rank(text):
    return sorted(list(FreqDist(text).items()), key=lambda x: x[1], reverse=True)
````

````python
def plotRankVsFreq(text):
    xs, ys = zip(*[(i, item[1]) for i, item in enumerate(rank(text))])
    
    fig, ax = plt.subplots(1, 1, figsize=(16, 8))
    ax.plot(xs, ys)
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.set_xlabel('Rank')
    ax.set_ylabel('Frequency')
    plt.show()

plotRankVsFreq(text2)
````

At the end with high ranks, the words occure very few times, where we at the end can see sudden drops in the frequency. These hard drops comes because a single occurence will make a difference.

````python
def generateRandomText(length = 1000):
    string = ''.join(random.choice('abcdefghijklmnopqrstuvxyz ') for _ in range(length))
    return string.split(' ')
````

````python
plotRankVsFreq(generateRandomText(int(1e+7)))
````
